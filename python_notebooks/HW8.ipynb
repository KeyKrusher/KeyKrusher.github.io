```python
import altair as alt

data_url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"

selection = alt.selection_single(on='click', fields=['Agency Name'], empty='none')

chart1 = alt.Chart(data_url).mark_bar().encode(
    x=alt.X('Agency Name:N', sort='-y', axis=alt.Axis(title="Institution")),
    y=alt.Y('count():Q', axis=alt.Axis(title="Number of Buildings")),
    tooltip=['Agency Name:N', 'count()'],
    color=alt.condition(selection, alt.value('steelblue'), alt.value('lightgray'))
).properties(
    width=600,
    height=400
).add_selection(
    selection
).interactive()

chart1
```

```python
# Load the data
data_url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"

# Define the second chart
chart2 = alt.Chart(data_url).transform_filter(
    alt.datum['Year Acquired'] > 1867
).mark_bar().encode(
    x=alt.X('Year Acquired:Q', bin=alt.Bin(maxbins=50), axis=alt.Axis(title="Year Acquired")),
    y=alt.Y('count():Q', axis=alt.Axis(title="Number of Buildings")),
    tooltip=['Year Acquired:Q', 'count()']
).properties(
    width=600,
    height=400
).interactive()

chart2
```

```python
# Save the visualizations to Jekyll project
jekyll_path = '../assets/json/'
chart1.properties(width='container').save(jekyll_path + "chart1.json")
chart2.properties(width='container').save(jekyll_path + "chart2.json")
```
